```python
import pandas as pd
```

```python
df = pd.read_csv("Visit-Nominal.csv", sep=",")
```

```python
df
```

```python
import numpy as np
```

```python
df.Visit_Again.size
```

```python
target_names = np.array([1,0])
```

```python
targets = np.array([])
```

```python
df.Home[df.Home == 'yes'] = 1
df.Home[df.Home == 'no'] = 0
df.Browsed[df.Browsed == 'yes'] = 1
df.Browsed[df.Browsed == 'no'] = 0
df.Searched[df.Searched == 'yes'] = 1
df.Searched[df.Searched == 'no'] = 0
df.Prod_A[df.Prod_A == 'yes'] = 1
df.Prod_A[df.Prod_A == 'no'] = 0
df.Prod_B[df.Prod_B == 'yes'] = 1
df.Prod_B[df.Prod_B == 'no'] = 0
df.Prod_C[df.Prod_C == 'yes'] = 1
df.Prod_C[df.Prod_C == 'no'] = 0
df.Visit_Again[df.Visit_Again == 'yes'] = 1
df.Visit_Again[df.Visit_Again == 'no'] = 0
```

```python
for i in range(df.Visit_Again.size):
    if df.Visit_Again[i] == 1:
        targets = np.append(targets,1)
    else:
        targets = np.append(targets,0)
```

```python
targets
```

```python
df['is_train'] = True
```

```python
for i in range(df.Visit_Again.size):
    if i<30:
        df['is_train'][i]=False
```

```python
df['Type'] = pd.Categorical.from_codes(targets, target_names)
```

```python
df['Targets'] = targets
```

```python
train = df[df['is_train']==True]
```

```python
test = df[df['is_train']==False]
```

```python
trainTargets = np.array(train['Targets']).astype(int)
```

```python
testTargets = np.array(test['Targets']).astype(int)
```

```python
features = df.columns[0:6]
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
y_gnb = gnb.fit(train[features], trainTargets).predict(test[features])
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
```

```python
y_mnb = mnb.fit(train[features], trainTargets).predict(test[features])
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
bnb = BernoulliNB()
```

```python
y_bnb = bnb.fit(train[features], trainTargets).predict(test[features])
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(y_gnb)
```

```python
print(accuracy_score(testTargets, y_gnb))
```

```python
print(y_mnb)
```

```python
print(accuracy_score(testTargets, y_mnb))
```

```python
print(y_bnb)
```

```python
print(accuracy_score(testTargets, y_bnb))
```
